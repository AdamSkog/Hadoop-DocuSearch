```python
import pyterrier as pt
import os
import pandas as pd
import warnings
warnings.filterwarnings("ignore") # To get rid of pyterrier warnings

if not pt.started():
  pt.init()
```

```python
base_path = os.path.abspath("../sites")

# Define the paths to the folders containing the downloaded files
folders = [
    os.path.join(base_path, "techcrunch.com"),
    os.path.join(base_path, "www.theverge.com"),
    os.path.join(base_path, "www.wired.com"),
]

# Collect files from all folders
all_files = []
for folder in folders:
    all_files.extend(pt.io.find_files(folder))

# Define the index path
index_path = os.path.join(base_path, "tech_news_index")

# Check if the index already exists
if os.path.exists(os.path.join(index_path, "data.properties")):
    index = pt.IndexFactory.of(index_path)
else:
    # Create a new index
    indexer = pt.FilesIndexer(index_path, meta={"docno": 20, "filename": 1024, "title": 1024}, meta_tags={"title": "title"})
    indexref = indexer.index(all_files)
    index = pt.IndexFactory.of(indexref)

# Print the collection statistics
print(index.getCollectionStatistics().toString())
```
